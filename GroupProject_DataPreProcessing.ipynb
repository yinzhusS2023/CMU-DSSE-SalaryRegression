```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

ngram_range = (1, 2)
min_df = 0.01
max_df = 0.8
max_features = 100

vectorizer = TfidfVectorizer(ngram_range=ngram_range, min_df=min_df, max_df=max_df, max_features=max_features)

# Get object columns
object_columns = [col for col in df3.columns if df3[col].dtype == 'object']

# Create an empty DataFrame to store the transformed columns
transformed_columns_df = pd.DataFrame(index=df3.index)

# Apply TfidfVectorizer to each object column and concatenate the result
for col in object_columns:
    # Drop any NaN values in the column to avoid errors during transformation
    non_null_df = df3.dropna(subset=[col])
    
    # Fit and transform the column using the TfidfVectorizer instance
    transformed_col = vectorizer.fit_transform(non_null_df[col])

    # Convert the sparse matrix to a dense DataFrame with appropriate column names
    transformed_col_df = pd.DataFrame(transformed_col.todense(), columns=[f"{col}_{i}" for i in range(transformed_col.shape[1])], index=non_null_df.index)

    # Concatenate the transformed column DataFrame to the main DataFrame
    transformed_columns_df = pd.concat([transformed_columns_df, transformed_col_df], axis=1)

# Drop the original object columns from the DataFrame
df3_without_objects = df3.drop(object_columns, axis=1)

# Concatenate the transformed columns DataFrame with the original DataFrame (excluding object columns)
df3_transformed = pd.concat([df3_without_objects, transformed_columns_df], axis=1)

df3_transformed.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df3_normalized = pd.DataFrame(scaler.fit_transform(df3_transformed), columns=df3_transformed.columns, index=df3_transformed.index)
df3_normalized.head()
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data, train_label, test_label = train_test_split(df3_normalized, salary, test_size=0.15, random_state=42)
print(train_data.shape)
print(train_label.shape)
print(test_data.shape)
print(test_label.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import BayesianRidge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


# Linear Regression
lr = LinearRegression()
lr.fit(train_data, train_label)
lr_predictions_training = lr.predict(train_data)
lr_predictions = lr.predict(test_data)
print("Training set: Linear Regression Root Mean Squared Error:", np.sqrt(mean_squared_error(train_label, lr_predictions_training)))
print("Training set: Linear Regression R2:", r2_score(train_label, lr_predictions_training))
print("Linear Regression Root Mean Squared Error:", np.sqrt(mean_squared_error(test_label, lr_predictions)))
print("Linear Regression R2:", r2_score(test_label, lr_predictions))
```

```python
# Decision Tree Regressor
dt = DecisionTreeRegressor(random_state=42)
dt.fit(train_data, train_label)
dt_predictions_training = dt.predict(train_data)
dt_predictions = dt.predict(test_data)
print("\n\nTraining set: Decision Tree Root Mean Squared Error:", np.sqrt(mean_squared_error(train_label, dt_predictions_training)))
print("Training set: Decision Tree R2", r2_score(train_label, dt_predictions_training))
print("Decision Tree Root Mean Squared Error:", np.sqrt(mean_squared_error(test_label, dt_predictions)))
print("Decision Tree R2", r2_score(test_label, dt_predictions))


# Neural Network Regression
nn = MLPRegressor(random_state=42)
nn.fit(train_data, train_label)
nn_predictions_training = nn.predict(train_data)
nn_predictions = nn.predict(test_data)
print("\n\nTraining set: Neural Network Root Mean Squared Error:", np.sqrt(mean_squared_error(train_label, nn_predictions_training)))
print("Training set: Neural Network R2", r2_score(train_label, nn_predictions_training))
print("Neural Network Root Mean Squared Error:", np.sqrt(mean_squared_error(test_label, nn_predictions)))
print("Neural Network R2", r2_score(test_label, nn_predictions))

# Bayesian Regression
br = BayesianRidge()
br.fit(train_data, train_label)
br_predictions_training = br.predict(train_data)
br_predictions = br.predict(test_data)
print("\n\nTraining set: Bayesian Regression Root Mean Squared Error:", np.sqrt(mean_squared_error(train_label, br_predictions_training)))
print("Training set: Bayesian Regression R2", r2_score(train_label, br_predictions_training))
print("Bayesian Regression Root Mean Squared Error:", np.sqrt(mean_squared_error(test_label, br_predictions)))
print("Bayesian Regression R2", r2_score(test_label, br_predictions))


# K-Nearest Neighbors Regression
knn = KNeighborsRegressor()
knn.fit(train_data, train_label)
knn_predictions_training = knn.predict(train_data)
knn_predictions = knn.predict(test_data)
print("\n\nTraining set: Bayesian Regression Root Mean Squared Error:", np.sqrt(mean_squared_error(train_label, br_predictions_training)))
print("Training set: Bayesian Regression R2", r2_score(train_label, br_predictions_training))
print("K-Nearest Neighbors Root Mean Squared Error:", np.sqrt(mean_squared_error(test_label, knn_predictions)))
print("Bayesian Regression R2", r2_score(test_label, knn_predictions))


# Gradient Boosting Regression
gbr = GradientBoostingRegressor(random_state=42)
gbr.fit(train_data, train_label)
gbr_predictions_training = gbr.predict(train_data)
gbr_predictions = gbr.predict(test_data)
print("\n\nTraining set: Gradient Boosting Root Mean Squared Error:", np.sqrt(mean_squared_error(train_label, gbr_predictions_training)))
print("Training set: Gradient Boosting R2", r2_score(train_label, gbr_predictions_training))
print("Gradient Boosting Root Mean Squared Error:", np.sqrt(mean_squared_error(test_label, gbr_predictions)))
print("Gradient Boosting R2", r2_score(test_label, gbr_predictions))
```
